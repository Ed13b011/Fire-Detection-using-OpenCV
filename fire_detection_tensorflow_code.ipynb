```python
# import opencv and numpy module
import numpy as np
import cv2

# create your own haarcascade classifier to detect fire in the video 
# For creating haarcascade classifier I follwed below steps:-
# 1. Use bulk images (I used 40 screenshots) of the input(Firemp4) video with different fire directions
# 2. Crop all images(563*288) for good accuracy and less time to train cascade classifier
# 3. Save these input images with folder name-> p. p for positive images
# 5. Next use around 200 (double the size of positive images )images for negative images
#(grayscale images gives better accuracy I used garyscale images of 300 for negative images)
 #6. Save negative(no fire images) images with folder name-> n. n for negative images 
# 4. Train these two folders (positive and negative images) in Cascade-Trainer-GUI
# 5. Output gives the haarcascade classifier for input video detection.

fire_haarcascade = cv2.CascadeClassifier('fire_detection_haarcascade_5.xml')

#fire_detection_haarcascade_5.xml contains the fire features
# xml file, input video and python code should be in the same folder while running the code

Test_Video = "Firemp4.mp4" 

# input video for testing

cap = cv2.VideoCapture(Test_Video)

# Get the Default resolutions of height and width to match the input video and output video
cap_width = int(cap.get(3))

cap_height = int(cap.get(4))

# Define the codec and create VideoWriter object
# fourcc for recording

out = cv2.VideoWriter('output_final.avi',cv2.VideoWriter_fourcc('M','J','P','G'), 10, (cap_width, cap_height))

while True:
    check, frame = cap.read()# reads frames from input video
    if check is True:
        
        gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY) # converting colored image to grayscale
        
        fire = fire_haarcascade.detectMultiScale(frame, scaleFactor = 1.2, minNeighbors = 5)    
        
        # detectMultiScale-> method to search for the fire rectangle coordinates 
        # scale factor-> decreses the shape of the value by 20% untill the fire is found
        # minNeighbors-> cascade classifier uses slides window proces to remove false positives and detect true positives
        
        for (x,y,w,h) in fire:
            
            x = x - 40 # resizig the rectangular box for fitting properly
            y = y - 40
            w = w + 30
            h = h + 30
            
            cv2.rectangle(frame,(x, y),(x + w, y + h),(255, 0, 0), 2) 
            # create rectangle and (255, 0, 0) for RGB value of rectangle outline and 2 is for width of the rectangle
            
            print('Fire is detected') # print alert message during fire detection
            
            out.write(frame)# write the frame for output video
    
    else:
        continue
        
    cv2.imshow('Result',frame)# shows output
                      
    # wait for key-> q to stop the program
    if cv2.waitKey(5) & 0xFF == ord('q'):
        break
        
print("came out of the loop")
#Release after run                      
cap.release()
out.release()
cv2.destroyAllWindows()
```
